### Import packages

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### Loading the datasets

```python
train_data = pd.read_csv('/content/drive/MyDrive/Labs/DI_lab/dataset_train_validation(P2-L2).csv')
train_data
```

```python
test_data = pd.read_csv('/content/drive/MyDrive/Labs/DI_lab/dataset_test(P2-L2).csv')
test_data
```

### Finding & replacing the *null values*

```python
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python
train_data.fillna(method='ffill', inplace=True)
test_data.fillna(method='ffill', inplace=True)
```

### Merging the datasets

```python
stockprice_data = pd.concat([train_data, test_data], axis=0)
stockprice_data
```

### Properties of Stockprice dataset

```python
stockprice_data.keys()
```

```python
print("The dimension:", stockprice_data.shape)
```

```python
print(stockprice_data.info())
```

```python
print(stockprice_data.describe())
```

### Splitting the data into *training* and *testing* sets

```python
stockprice_data = stockprice_data['close'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
combined_data = scaler.fit_transform(stockprice_data)
```

```python
train_data_len = len(train_data)
train_data = stockprice_data[:train_data_len]
test_data = stockprice_data[train_data_len:]
```

### Training the model using the ***Random Forest model***

```python
def train_random_forest_model(train_data):
    X_train = np.arange(1, len(train_data) + 1).reshape(-1, 1)
    y_train = train_data.flatten()

    # Handle missing values in y_train
    y_train = pd.Series(y_train).fillna(pd.Series(y_train).mean()).values

    model = RandomForestRegressor(n_estimators=100)
    model.fit(X_train, y_train)

    return model

rf_model = train_random_forest_model(train_data)
```

### Evaluation of model predictions

```python
X_test = np.arange(train_data_len, train_data_len + len(test_data) - 1).reshape(-1, 1)
rf_predictions = rf_model.predict(X_test)
```

```python
y_test = scaler.inverse_transform(test_data[:-1])
y_test = y_test.flatten()
```

### Calculating **RMSE** for *Random Forest* model

```python
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
print("Random Forest RMSE:", rf_rmse)
```
